```python
x = np.arange(np.pi*4)
sin = np.sin(x)
cos = np.cos(x)
# sin(x)
plt.figure(figsize=(12, 6))
plt.plot(linewidth=1.5,
         linestyle="-.",
         color="#8E44AD",
         alpha=0.7,
         marker="h")
# cos(x)
plt.plot(linewidth=1.5,
         linestyle=":",
         color="c",
         alpha=0.7,
         marker="v")

plt.title("graph of the sin(x) and cos(x) functions")
plt.xlabel("pi")

plt.legend(["sin(x)", "cos(x)"], loc=0, fontsize="x-large")
plt.grid(color="tomato", alpha=0.5, linestyle="--")
plt.show()
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
sin = np.sin(x) # sin(x)
cos = np.cos(x) # cos(x)

plt.figure(figsize=(12,6))
# sin(x) grafigi
plt.plot(x, sin,
         linewidth=1.5,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.7,
         marker='h')

# cos(x) grafigi
plt.plot(x, cos,
         linewidth=1.5,
         linestyle=':',
         color='c',
         alpha=0.7,
         marker='v')

# Grafik nomi:
plt.title('sin(x) va cos(x) funksiya grafigi')
# x-o'q nomi:
plt.xlabel("pi")
# Har bir chiziq haqida ma'lumot:
plt.legend(['sin(x)','cos(x)'], loc=0, fontsize='x-large')
plt.grid(color='tomato', alpha=0.5, linestyle='--') # setkani o'zgartirish
plt.show()
```

# heatmap

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv", index_col=0, parse_dates=[0], usecols=[0,1])
df.head()
```

```python
type(df.index)
```

```python
df["month"] = [i.month for i in df.index]
df.head()
```

```python
df["year"] = [i.year for i in df.index]
df.head()
```

```python
df2 = df.groupby(["month", "year"]).max()
df2
```

```python
df2.unstack(level=0)
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df2.unstack(level=0), cmap="YlGnBu", linewidths=1, square=True)
plt.show()
```

# distribution

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.github.com/anvarnarz/praktikum_datasets/main/merc.csv")
df = df[df.year>=2020]
df = df[df.transmission != "Other"]
df = df[df.fuelType != "Other"]
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df, x="price")
plt.show()
```

```python
sns.displot(data=df, x="price",hue="fuelType", height=5, aspect=2)
plt.show()
```

#kde

```python
sns.displot(data=df, x="price", kind="kde", height=5, aspect=2)
plt.show()
```

```python
sns.displot(data=df, x="price", hue="transmission", kde=True, height=5, aspect=2)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data=df, x="price", hue="fuelType", shade=True)
plt.show()
```

# CDE

```python
sns.displot(data=df, x="price", kind="ecdf", height=5, aspect=2)
plt.xticks(rotation=60)
plt.yticks(rotation=60)
plt.grid(which="both") # major , minor, both
plt.minorticks_on()
plt.show()
```

```python
sns.displot(data=df, x="price",  col="transmission", height=5, aspect=1)
plt.grid(which="minor")
plt.show()
```

```python
sns.displot(data=df, x="price", y="year", kind="kde", height=5, aspect=2)
plt.xticks(rotation=90)
plt.show()
```

# subplot

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.github.com/anvarnarz/praktikum_datasets/main/merc.csv")
df = df[df.year>=2020]
df = df[df.transmission != "Other"]
df = df[df.fuelType != "Other"]
df.head()
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
sin = np.sin(x) # sin(x)
cos = np.cos(x) # cos(x)

fig1 = plt.figure(figsize=(12,6))
plt.subplot(2,1,1)

plt.plot(x, sin,
         linewidth=1.5,
         linestyle="-.",
         color="#8E44AD",
         alpha=0.7,
         marker="h")
plt.title("sin(x) function graph")
plt.grid(color="tomato", alpha=0.5, linestyle="--")

plt.subplot(2,1,2)
plt.plot(x, cos,
         linewidth=1.5,
         linestyle=":",
         color="c",
         alpha=0.7,
         marker="v")
plt.title("cos(x) function graph")
plt.grid(color="tomato", alpha=0.5, linestyle="--")

plt.show()

```

```python
sns.histplot(data=df, x="price")
plt.show()
```

```python
sns.scatterplot(data=df, x="price", y="year", s=100)
plt.show()
```

```python
fig, ax = plt.subplots(2,1, figsize=(10, 10))
sns.histplot(ax= ax[0], data=df, x="price")
sns.scatterplot(ax=ax[1], data=df, x="price", y="mpg", s=100)
plt.show()
```
